```python
# elbow plot - for finding the best value of k

from sklearn.cluster import KMeans
import pandas as pd # Import pandas for DataFrame manipulation
import matplotlib.pyplot as plt

# Load or create your DataFrame here
df=pd.read_csv('Iris (1).csv')

# Drop non-numerical columns before fitting KMeans
X = df.drop(columns=['Id', 'Species']) # Assuming 'Id' is also non-numerical

sse=[]
k_range=range(1,10)
for k in k_range:
  km=KMeans(n_clusters=k)
  km.fit(X) # Fit KMeans on numerical data only
  sse.append(km.inertia_) # use inertia_attribute

plt.title('the elbow plot')
plt.xlabel('k')
plt.ylabel('sum of squared error')
plt.plot(k_range,sse)
```

the elbow plot shows that the best value of k is 3.

```python
# kmeans clustering
kmeans = KMeans(n_clusters=3, random_state=0, max_iter = 300, init = 'k-means++', n_init=10)
y_kmeans = kmeans.fit_predict(X)
```

```python
# predict outputs

y_kmeans
```

```python
# centoids

kmeans.cluster_centers_
```

```python
# visualisation based on petal length and petal width

plt.scatter(X[y_kmeans == 0]['PetalLengthCm'], X[y_kmeans == 0]['PetalWidthCm'], s = 100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1]['PetalLengthCm'], X[y_kmeans == 1]['PetalWidthCm'], s = 100, c = 'blue', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 2]['PetalLengthCm'], X[y_kmeans == 2]['PetalWidthCm'], s = 100, c = 'green', label = 'Iris-setosa')

# plotting centroids
# Use marker instead of markers
plt.scatter(kmeans.cluster_centers_[:,2], kmeans.cluster_centers_[:,3], marker='*', s = 100, c = 'yellow', label = 'Centroids')

plt.title('clustering based on petal features')
plt.xlabel('PetalLengthCm')
plt.ylabel('PetalWidthCm')
plt.legend()
```

```python
# Visualization based on sepal length and sepal width

plt.scatter(X[y_kmeans == 0]['SepalLengthCm'], X[y_kmeans == 0]['SepalWidthCm'], s =100, c = 'red', label = 'Iris-setosa')
plt.scatter(X[y_kmeans == 1]['SepalLengthCm'], X[y_kmeans == 1]['SepalWidthCm'], s =100, c = 'blue', label = 'Iris versicolor')
plt.scatter(X[y_kmeans == 2]['SepalLengthCm'], X[y_kmeans == 2]['SepalWidthCm'], s =100, c = 'green', label = 'Iris virginica')

# plotting centroids

plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], marker='*', s = 100, c = 'yellow', label = 'Centroids')

plt.title('clustering based on sepal features')
plt.xlabel('SepalLengthCm')
plt.ylabel('SepalWidthCm')
plt.legend()
```

This K-means clustering analysis effectively identifies the inherent structure of the Iris dataset and provides a clear visualization of the differences between the three Iris species. This type of analysis can be useful for various applications, such as species classification, customer segmentation, or anomaly detection.
